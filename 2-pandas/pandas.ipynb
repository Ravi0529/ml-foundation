```python
df = pd.DataFrame({
    'name': ['alfred', 'batman', 'catwoman'],
    'toy': [np.nan, 'batmobile', 'bullwipe'],
    'born': [pd.NaT, pd.Timestamp('1940-04-25'), pd.NaT]
})
df.head()
```

```python
# drop the column if any is NaN, NaT, etc
df.dropna()
```

```python
# drop only if the whole column is NaN, NaT, etc.
df.dropna(how = "all")
```

```python
# df.loc[0, "name"] = "catwoman"
df.iloc[0, 0] = "catwoman"
df
```

```python
df.drop_duplicates(subset = ["name"])
```

```python
df.drop_duplicates(subset = ["name"], keep = "last")
```
