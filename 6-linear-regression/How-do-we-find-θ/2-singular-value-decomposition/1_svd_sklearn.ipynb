## Singular Value Decomposition (SVD)

### Why SVD is better: The complexity O(n²) is better than O(n³). More importantly, SVD uses the Moore-Penrose pseudoinverse `X⁺`, which is always defined. This means it can handle cases where the Normal Equation fails (e.g., if XᵀX is not invertible due to redundant features).


## Method A: Using Scikit-Learn (Easiest)


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# generating linear data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.rand(100, 1)
```

```python
# create and train the model
lin_reg = LinearRegression()
lin_reg.fit(X, y)  # sklearn automatically adds the bias term

# get the optimized parameters
print("\nOptimized parameters")
print(f"Intercept (θ0): {lin_reg.intercept_[0]:.6f}")
print(f"Coefficient (θ1): {lin_reg.coef_[0][0]:.6f}")
```

```python
# make predictions
X_new = np.array([[0], [2]])
y_predict = lin_reg.predict(X_new)

print("\nPredictions for new data")
for i in range(len(X_new)):
    print(f"x = {X_new[i][0]:.1f} -> predicted y = {y_predict[i][0]:.6f}")
```

```python
plt.figure(figsize=(10, 6))
plt.plot(X, y, "b.", label="Training Data", alpha=0.8)
plt.plot(X_new, y_predict, "r-", label="Regression Line", linewidth=2)
plt.plot(X_new, y_predict, "ro", label="Predictions", markersize=8)
plt.xlabel("Feature X₁", fontsize=14)
plt.ylabel("Target y", fontsize=14)
plt.title("Linear Regression with Scikit-Learn (SVD)", fontsize=16)
plt.legend(loc="upper left")
plt.grid(True)
plt.axis([0, 2, 0, 15])
plt.show()
```
