## Adding labels and title

```python
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("FOR BUYING A HOUSE")
plt.plot([1,2,3,4,6], [3,5,7,3,6])
```

```python
from matplotlib import style
# style.use("ggplot")
style.use("dark_background")
x1 = [1,5,7]
y1 = [23,45,67]
x2 = [4,6,8]
y2 = [14,25,57]
plt.plot(x1,y1, label = 'first', linewidth = 4)
plt.plot(x2,y2, label = 'second', linewidth = 4)
plt.legend(loc = "upper left")
plt.show()

# https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html
# refer this able link for more styles
```

```python
style.use("ggplot")
x1 = [1,5,7]
y1 = [23,45,67]
x2 = [4,6,8]
y2 = [14,25,57]
plt.scatter(x1, y1, label = "first", linewidth = 6) # width of the dot
plt.scatter(x2, y2, label = "second", linewidth = 6)
plt.legend(loc = "upper left")
plt.show()
```

```python
x = [1,2,4,5,6,3,7,3,1,2,2]
plt.hist(x, label = "col")
plt.legend(loc = "upper right")
plt.show()
```

```python
x = ['python','c++','java']
y = [2,4,6]
z = [1,3,5]
c = ['r', 'g', 'b']
plt.bar(x,y, width = 0.3, color = c, edgecolor = 'black', linewidth = 3, linestyle = ":", alpha = 0.4)
plt.bar(x,z, width = 0.3, color = 'y', edgecolor = 'black', linewidth = 3)
plt.show()
```

```python
x = [10, 20, 30, 40]
y = ["python", "java", "cpp", "go"]
ex = [0.3, 0.0, 0.0, 0.0]
plt.pie(x, labels = y, explode = ex, autopct = "%0.1f%%")
plt.legend(loc = 3)
plt.show()

# loc --> 1 = upper right, 2 = upper left, 3 = lower left, 4 = lower right
```

```python
x = [10, 20, 30, 40]
y = [1, 2, 3, 4]
c = ["r", "b", "g", "y"]
plt.pie(x, labels = y, colors = c, radius = 1) # radius of the outer circle
plt.pie([1], colors = "w", radius = 0.5) # for inner circle
plt.title("Pie chart (Donut)")
plt.legend(loc = 4)
plt.show()
```

```python
style.use("ggplot")
x = [1,2,3,4,5]
y = [2,3,4,2,3]
plt.stem(x, y, linefmt = ":", markerfmt = "bo", basefmt = "g", label = "python")
plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5]
area1 = [2,4,5,3,6]
area2 = [1,3,5,4,2]
area3 = [1,3,5,6,3]
l = ['area1', 'area2', 'area3']
plt.stackplot(x, area1, area2, area3, labels = l)
plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,2,3,5,4]
plt.step(x,y, color = 'r', marker = "o", ms = 10, mfc = 'g', label = 'python')
plt.legend(loc = 2)
plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])
plt.plot(x, y, color = "r")
plt.fill_between(x, y, color = "g", where = (x >= 2) & (x <= 4), alpha = 0.3)
plt.show()
```

```python
x = [10,20,30,40,50,60,70]
plt.boxplot(x)
plt.show()
```

```python
x = [10,20,30,40,50,100] # 100 is the outlier
# plt.boxplot(x)
plt.boxplot(x, notch = True)
plt.show()
```

```python
x = [10,20,30,40,50,60,70,120]
plt.boxplot(x, sym = "g")
plt.show()
```

```python
x = [0,30,40,50,55,60,80,200,210]
plt.boxplot(x, patch_artist = True, showmeans = True, sym = "b", vert = False)
plt.show()
```
